```
print ("Aplikasi Hitung Rumus Geometri")
print ("Pilih Jenis")
print ("[1.] Bangun Datar")
print ("[2.] Bangun Ruang")
pilih = input("pilih : ")

##Bagian ini untuk Bangun Datar
if pilih == '1':
  print ("Pilih Bangun Datar")
  print ("[1.] Persegi")
  print ("[2.] Persegi Panjang")
  print ("[3.] Lingkaran")
  print ("[4.] Segitiga")
  print ("[5.] Trapesium")
  pilih = input("pilih : ")

  if pilih == '1': ##Persegi
    print ("Pilih Rumus Bangun Datar")
    print ("[1.] Luas")
    print ("[2.] Keliling")
    pilih = input("pilih : ")

    if pilih == '1': ##Luas Persegi
      print ("Rumus Luas Dari Bangun Persegi")
      print ("Rumusnya: sisi x sisi")

      s = float(input("Masukkan Sisi: "))
      luas = s * s

      print("Hasil Perhitungannnya = ",luas)

    elif pilih == '2': ##Keliling Persegi
      print ("Rumus keliling Dari Bangun Persegi")
      print ("Rumusnya: 4 x sisi")

      s = float(input("Masukkan Sisi: "))
      keliling = 4 * s

      print("Hasil Perhitungannnya = ",keliling)
      ##DONE
  
  elif pilih == '2': ##Persegi Panjang
    print ("Pilih Rumus Bangun Datar")
    print ("[1.] Luas")
    print ("[2.] Keliling")
    pilih = input("pilih : ")
    
    if pilih == '1': ##Luas Persegi Panjang
      print ("Rumus Luas Dari Bangun Persegi Panjang")
      print ("Rumusnya: panjang x lebar")

      panjang = float(input("Masukan Panjang: "))
      lebar = float(input("Masukan Lebar: "))

      luas = panjang * lebar

      print("Hasil Perhitungannnya = ",luas)

    elif pilih == '2': ##Keliling Persegi Panjang
      print ("Rumus Keliling Dari Bangun Persegi Panjang")
      print ("Rumusnya: 2 x (panjang + lebar)")

      panjang = float(input("Masukan Panjang: "))
      lebar = float(input("Masukan Lebar: "))

      keliling = 2 * (panjang + lebar)

      print ("Hasil Perhitungannnya = ",keliling)
      ##DONE

  elif pilih == '3': #Lingkaran
    print ("Pilih Rumus Bangun Datar")
    print ("[1.] Luas")
    print ("[2.] Keliling")
    pilih = input("pilih : ")

    if pilih == '1': ##Luas Lingkaran
      print ("Rumus Luas Dari Bangun Lingkaran")
      print ("Rumusnya: π × r²")

      import math
      r = float(input("Masukan Jari-jari : "))

      luas = math.pi * (r * r)

      print("Hasil Perhitungannnya = ",luas)

    elif pilih == '2': ##Keliling Lingkaran
      print ("Rumus Keliling Dari Bangun Lingkaran")
      print ("Rumusnya: 2 x π × r")

      import math
      r = float(input("Masukan Jari-jari : "))

      keliling = 2 * math.pi * r

      print("Hasil Perhitungannnya = ",keliling)
      ##DONE

  elif pilih == '4': #Segitiga
    print ("Pilih Rumus Bangun Datar")
    print ("[1.] Luas")
    print ("[2.] Keliling")
    pilih = input("pilih : ")

    if pilih == '1': ##Luas Segitiga
      print ("Rumus Luas Dari Bangun Segitiga")
      print ("Rumusnya: alas x tinggi / 2")

      alas = float(input("Masukan Alas: "))
      tinggi = float(input("Masukan Tinggi: "))

      luas = 0.5 * alas * tinggi

      print("Hasil Perhitungannnya = %0.2f"%luas)
    
    elif pilih == '2': ##Keliling Segitiga
      print ("Rumus Keliling Dari Bangun Segitiga")
      print ("Rumusnya: S1 + S2 + S3")

      S1 = int(input("Masukan S1: "))
      S2 = int(input("Masukan S2: "))
      S3 = int(input("Masukan S3: "))

      keliling = S1 + S2 + S3

      print("Hasil Perhitungannnya = ",keliling)
      ##DONE

  elif pilih == '5': #Trapesium
    print ("Pilih Rumus Bangun Datar")
    print ("[1.] Luas")
    print ("[2.] Keliling")
    pilih = input("pilih : ")

    if pilih == '1': ##Luas Trapesium
      print ("Rumus Luas Dari Bangun Trapesium")
      print ("Rumusnya: (S1 + S2) x T / 2")

      S1 = float(input("Masukkan Sisi atas: "))
      S2 = float(input("Masukkan Sisi Bawah: "))
      tinggi = float(input("masukkan tinggi trapesium : "))

      luas = (S1 + S2) * tinggi/2
      
      print("Hasil Perhitungannnya = ",luas,"cm2")
    
    elif pilih == '2': ##Keliling Trapesium
      print ("Rumus Keliling Dari Bangun Trapesium")
      print ("Rumusnya: S1 + S2 + S3 + S4")

      S1 = int(input("Masukan S1: ")) ##S1-S4 itu artinya sisi/rusuk
      S2 = int(input("Masukan S2: "))
      S3 = int(input("Masukan S3: "))
      S4 = int(input("Masukan S4: "))

      keliling = S1 + S2 + S3 + S4

      print("Hasil Perhitungannnya = ",keliling)
      ##DONE

##Bagian ini untuk Bangun Ruang
elif pilih == '2':
  print ("Pilih Bangun Ruang")
  print ("[1.] Kubus")
  print ("[2.] Balok")
  print ("[3.] Bola")
  print ("[4.] Kerucut")
  print ("[5.] Silinder/Tabung")
  pilih = input("pilih : ")

  if pilih == '1': ##Kubus
    print ("Pilih Rumus Bangun Datar")
    print ("[1.] Volume")
    print ("[2.] Luas Permukaan")
    pilih = input("pilih : ")

    if pilih == '1': ##Volume Kubus
      print ("Rumus Volume Dari Kubus")
      print ("Rumusnya: sisi x sisi x sisi")

      sisi = float(input("Masukkan Sisi : "))
      hasil = sisi * sisi * sisi

      print("Hasil Perhitungannnya = ",hasil)

    elif pilih == '2': ##Luas Permukaan Kubus
      print ("Rumus Luas Permukaan Dari Kubus")
      print ("Rumusnya: 6 x sisi x sisi")

      sisi = float(input("Masukkan Panjang Sisi : "))
      hasil = 6 * sisi * sisi

      print("Hasil Perhitungannnya = ",hasil)
      ##DONE
  
  elif pilih == '2': ##Balok
    print ("Pilih Rumus Bangun Datar")
    print ("[1.] Volume")
    print ("[2.] Luas Permukaan")
    pilih = input("pilih : ")

    if pilih == '1': ##Volume Balok
      print ("Rumus Volume Dari Balok")
      print ("Rumusnya: p × l × t")

      p = int(input("Masukan panjang balok: "))
      l = int(input("Masukan lebar balok: "))
      t = int(input("Masukan tinggi balok: "))

      hasil = p * l * t

      print("Hasil Perhitungannnya = ",hasil)

    elif pilih == '2': ##Luas Permukaan Balok
      print ("Rumus Volume Dari Balok")
      print ("Rumusnya: 2 × (p.l + p.t +l.t)")

      p = int(input("Masukan panjang balok: "))
      l = int(input("Masukan lebar balok: "))
      t = int(input("Masukan tinggi balok: "))

      hasil = 2 * ((p*l) + (p*t) + (l*t))

      print("Hasil Perhitungannnya = ",hasil)
      ##DONE

  elif pilih == '3': ##Bola
    print ("Pilih Rumus Bangun Datar")
    print ("[1.] Volume")
    print ("[2.] Luas Permukaan")
    pilih = input("pilih : ")

    if pilih == '1': ##Volume Bola
      print ("Rumus Volume Dari Bola")
      print ("Rumusnya: 4/3 × π × r³")

      jari = int(input("Masukan jari-jari Bola : "))

      phi = (22/7)
      hasil = int((4/3)*phi*(jari*jari*jari))

      print("Hasil Perhitungannnya = ",hasil)

    elif pilih == '2': ##Luas Permukaan Bola
      print ("Rumus Luas Permukaan Dari Bola")
      print ("Rumusnya: 4 × π × r²")

      jari = int(input("Masukan jari-jari Bola : "))

      phi = (22/7)
      hasil = int(4*phi*(jari*jari))

      print("Hasil Perhitungannnya = ",hasil)
      ##DONE

  elif pilih == '4': ##Kerucut
    print ("Pilih Rumus Bangun Datar")
    print ("[1.] Volume")
    print ("[2.] Luas Permukaan")
    pilih = input("pilih : ")

    if pilih == '1': ##Volume Kerucut
      print ("Rumus Volume Dari Kerucut")
      print ("Rumusnya: 1/3 π x r x r x t")

      jari = int(input("Masukan jari-jari Kerucut : "))
      tinggi = int(input("Masukan Tinggi Kerucut : "))

      phi = (22/7)
      hasil = 1/3*(phi*jari*jari*tinggi)

      print("Hasil Perhitungannnya = ",hasil)

    elif pilih == '2': ##Luas Permukaan Kerucut
      print ("Rumus Luas Permukaan Dari Kerucut")
      print ("Rumusnya: luas alas + luas selimut")

      alas = int(input("Masukan Luas Alas Kerucut : "))
      selimut = int(input("Masukan Luas Selimut Kerucut : "))

      phi = (22/7)
      hasil = alas + selimut

      print("Hasil Perhitungannnya = ",hasil)
      ##DONE

  elif pilih == '5': ##Silinder/Tabung
    print ("Pilih Rumus Bangun Datar")
    print ("[1.] Volume")
    print ("[2.] Luas Permukaan")
    pilih = input("pilih : ")

    if pilih == '1': ##Volume Silinder/Tabung
      print ("Rumus Volume Dari Silinder/Tabung")
      print ("Rumusnya: π × r × r × t")

      tinggi = int(input("Masukan Tinggi Tabung : "))
      jari = int(input("Masukan Jari-jari Tabung : "))

      phi=22/7
      hasil = int((phi*(jari*jari))*tinggi)

      print("Hasil Perhitungannnya = ",hasil)

    elif pilih == '2': ##Luas Permukaan Silinder/Tabung
      print ("Rumus Luas Permukaan Dari Silinder/Tabung")
      print ("Rumusnya:  2 × π × r × (r + t)")

      tinggi = int(input("Masukan Tinggi Tabung : "))
      jari = int(input("Masukan Jari-jari Tabung : "))

      phi=22/7
      hasil = int(2*phi*jari*(tinggi+jari))

      print("Hasil Perhitungannnya = ",hasil)
      ##DONE
```
